```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("doctor-strange-2-benedict-cumberbatch.jpg", 0)

plt.imshow(img,'gray')
plt.show()
print(img.shape)
```

```python
kernel=[]

kernel.append([1,1,1])
kernel.append([1,-8,1])
kernel.append([1,1,1])

kernel=np.array(kernel)

print(kernel)
```

```python
new_img=np.zeros((img.shape[0]-kernel.shape[0]+1,img.shape[1]-kernel.shape[1]+1))

print(new_img.shape)
```

```python
for ix in range(new_img.shape[0]):
    for jx in range(new_img.shape[0]):
        
        img_patch=img[ix:ix+kernel.shape[0],jx:jx+kernel.shape[1]]
        
        product=img_patch*kernel
        
        new_img[ix][jx]=max(0,product.sum())
        

        
```

```python
plt.imshow(new_img.astype(np.uint8),cmap='gray')
plt.show()
```
