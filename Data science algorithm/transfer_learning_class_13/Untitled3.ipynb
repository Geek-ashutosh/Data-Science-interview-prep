```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.layers import Input,Flatten,Dense,Dropout,Conv2D,GlobalAveragePooling2D
from keras.optimizers import Adam
from keras.activations import softmax,relu
from keras.models import Model
from keras.preprocessing import image
from keras.utils import np_utils
from keras.applications.inception_resnet_v2 import InceptionResNetV2
import os
```

```python
dirs=os.listdir("./data")
print(dirs)
```

```python
path="./data/"
```

```python
images=[]
labels=[]
```

```python
labels_dict={'horses':0,'cats':1,'dogs':2,'Humans':3}
```

```python
for ix in dirs:
    class_path=path + ix +"/"
    img_name=os.listdir(class_path)
    
    for im in img_name:
        
        im=image.load_img(class_path +im,target_size=(224,224,3))
        im_array=image.img_to_array(im)
        
        images.append(im_array)
        labels.append(labels_dict[ix])
        
```

```python
X_train=np.array(images)
Y_train=np.array(labels)
Y_train=np_utils.to_categorical(Y_train)
print(X_train.shape,Y_train.shape)
```

```python
res_model=InceptionResNetV2(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
res_model.summary()
```

```python
c1=Conv2D(1,(3,3),activation='relu')(res_model.output)
f1=Flatten()(c1)
f2=Dense(4,activation='softmax')(f1)

model=Model(input=res_model.inputs,outputs=c1)
model.summary()
```

```python
avg = GlobalAveragePooling2D()(res_model.output)
fc1 = Dense(256, activation='relu')(avg)
fc2 = Dense(4, activation='softmax')(f2)

model = Model(inputs=res_model.inputs, outputs=f2)
model.summary()
```

```python
for ix,layers in enumerate(model.layers):
    print(ix,layers)
    
```

```python
for ix in range(171):
    model.layers[ix].trainable=False
```

```python
model.summary()
```

```python
adam=Adam(lr=0.0001)
```

```python
model.compile(optimizer=adam,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
hist = model.fit(X_train,Y_train,epochs=5,shuffle=True,batch_size=64,validation_split=0.20)
```

```python
plt.figure(0)
plt.plot(hist.history['acc'],'r')
plt.plot(hist.history['loss'],'b')
```
