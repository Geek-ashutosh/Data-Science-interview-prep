```python
import numpy as np
import gensim
import pandas as pd
import sys
```

```python
train_orders=pd.read_csv('./order_products__train.csv')
products=pd.read_csv('./products.csv')
```

```python
train_orders.head(10)
```

```python
products.tail(10)
```

```python
train=pd.merge(train_orders,products,how='left',on='product_id')
```

```python
train.head
```

```python
train_products=train.groupby('order_id').apply(lambda order :order["product_name"].tolist())
```

```python
train_products.head
```

```python
sentences=train_products.values
```

```python
sentences[5]
```

```python
model_word2vec=gensim.models.Word2Vec(sentences,window=5,sg=1,workers=4,min_count=2)
```

```python
model_word2vec.wv.most_similar(positive=["Roasted Turkey", "Clementines"], topn=3)
```
