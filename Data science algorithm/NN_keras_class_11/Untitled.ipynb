```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential,Model
from keras.optimizers import Adam,SGD
from keras.layers import Input,Dense
from keras.utils import np_utils
from keras.activations import tanh,sigmoid,softmax,relu
```

```python
data = pd.read_csv("./fashion-mnist_test.csv")
```

```python
x=data.values
X=x[:,1:]
Y=x[:,0]

X=X/255.0

Y=np_utils.to_categorical(Y)
print (X.shape,Y.shape)

X_train=X[:8000,:]

Y_train=X[:8000,:]

X_test=X[8000:,]
Y_test=X[8000:,]
```

```python
model = Sequential()

model.add(Dense(256, activation='tanh', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
adam=Adam(decay=0.005)
```

```python
model.compile(optimizer=adam,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,Y_train,epochs=10,batch_size=200,shuffle=True,validation_data=(X_test,Y_test))
```
