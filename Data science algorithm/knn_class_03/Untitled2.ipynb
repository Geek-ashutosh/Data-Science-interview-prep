```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
X, y = make_blobs(n_samples=1000, n_features=2, center_box=(-1,1),centers=3, random_state=2)
```

```python
q_point = np.array([-1.0,-3.0])
```

```python
plt.scatter(X[:,0], X[:,1],c=y)
plt.scatter(q_point[0], q_point[1], c='red')
plt.show()
```
